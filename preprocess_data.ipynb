```python
import pandas as pd

rental_affordability = pd.read_csv('rental_affordability.csv') 
rental_affordability['departement_code'] = rental_affordability['code_commune'].apply(lambda x: str(x)[:2])
rental_affordability.to_csv('rental_affordability_departement.csv', index=False)
```

```python
rental_affordability['departement_code'].value_counts()
```

```python
rental_affordability.head(5)
```

```python
# 聚合求二分位数，中位数，四分位数 根据region_code
rental_affordability['median'] = rental_affordability.groupby('departement_code')['price_euro_m2'].transform('median')
rental_affordability['q1'] = rental_affordability.groupby('departement_code')['price_euro_m2'].transform(lambda x: x.quantile(0.25))
rental_affordability['q3'] = rental_affordability.groupby('departement_code')['price_euro_m2'].transform(lambda x: x.quantile(0.75))
# 最大值和最小值
rental_affordability['max'] = rental_affordability.groupby('departement_code')['price_euro_m2'].transform('max')
rental_affordability['min'] = rental_affordability.groupby('departement_code')['price_euro_m2'].transform('min')
rental_affordability
```

```python
# 只要region以及中位数，二分位数，四分位数的数据
departement_rental_data = rental_affordability[['departement_code', 'median', 'q1', 'q3','max','min']].drop_duplicates().sort_values(by='departement_code').reset_index(drop=True)
```
